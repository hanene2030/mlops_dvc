```python
from sklearn import datasets
```

```python
features, output, coef = datasets.make_regression(n_samples = 1000, n_features = 4,
                                n_informative = 4, n_targets = 1,
                                noise = 0.0, coef = True,random_state=32)
```

```python
df = pd.DataFrame(features+4, columns=['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4'])
df['Target'] = output




#print(pd.DataFrame(coef, columns=['True Coefficient Values']))
```
